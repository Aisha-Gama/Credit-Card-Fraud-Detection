```python
import pandas as pd
```

```python
#load the data
credit_card_df = pd.read_csv('creditcard.csv')
```

```python
#check for distribution of the target
credit_card_df['Class'].value_counts()
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler

credit_card_df['Amount_scaled'] = StandardScaler().fit_transform(credit_card_df['Amount'].values.reshape(-1,1))

credit_card_df = credit_card_df.drop(['Amount', 'Time'], axis=1)

credit_card_df.head()
```

```python
#fix by over sampling the fradulent class using SMOTE to help catch the fraud
from imblearn.over_sampling import SMOTE

# Separate the features (X) from the target variable (y)
X = credit_card_df.drop('Class', axis=1)
y = credit_card_df['Class']

# Apply SMOTE to balance the classes
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the new balance of classes
pd.Series(y_resampled).value_counts()
```

```python
#split data to test the model on new data it hasn't seen before to see how well it performs
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Check the shape of the training and testing data
X_train.shape, X_test.shape
```

```python
#train the model to recognize fraud using Logistic Regressior as its a binary classification
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression()

# Train the model with the training data
model.fit(X_train, y_train)
```

```python
#test the trained model the testing data to evaluate its performance using accuracy, precision, recall, and F1-score
from sklearn.metrics import classification_report, accuracy_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model's performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
#visualize the result to show how the precision, recall, and F1-scores compare between the two classes
import matplotlib.pyplot as plt
import numpy as np

# Values for precision, recall, and f1-score from the classification report
metrics = ['Precision', 'Recall', 'F1-Score']
class_0 = [0.92, 0.98, 0.95]  # Values for Class 0 (non-fraudulent)
class_1 = [0.97, 0.92, 0.95]  # Values for Class 1 (fraudulent)

# Plotting the bar chart
bar_width = 0.3  # Width of the bars
index = np.arange(len(metrics))  # The x locations for the groups

# Create subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Bars for Class 0 and Class 1
bars_class_0 = ax.bar(index, class_0, bar_width, label='Class 0 (Non-fraudulent)', color='purple')
bars_class_1 = ax.bar(index + bar_width, class_1, bar_width, label='Class 1 (Fraudulent)', color='orange')

# Add labels and titles
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Classification Report: Precision, Recall, and F1-Score')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(metrics)
ax.legend()

# Display the bar chart
plt.show()
```
